```python
import pandas as pd
import numpy as np
import yfinance as yf

from data_collector import DataCollector
```

```python
companies_df = pd.read_csv(r'data\sp500-with-gics.csv')
companies_df
```

```python

tickers = list(companies_df["Ticker"])
tickers
```

```python
data_collector = DataCollector()
data_collector.set_tickers(tickers)
```

```python
data_collector.download_data("1999-12-01", "2024-01-01", "1d")
```

```python
raw_data = data_collector.get_raw_data()
raw_data.info()
adj_close_df = raw_data["Adj Close"]
```

```python
adj_close_df
```

```python
nan_count_dict = {}

for column in adj_close_df.columns:
    nan_count_dict[column] = len(adj_close_df[adj_close_df[column].isna()].index.tolist())
```

```python
nan_count_dict
```

```python
valid_columns = [column for column, nan_count in nan_count_dict.items() if nan_count == 0]
filtered_df = adj_close_df[valid_columns]

quarterly_df = filtered_df.resample('Q').first()

quarterly_return_rates_df = quarterly_df.pct_change()
quarterly_return_rates_df.dropna(inplace=True) # dropping first row as from the second one there are non-nan values

quarterly_return_rates_df.to_csv(r'./data/quarterly_return_rates.csv')

quarterly_return_rates_df
```
