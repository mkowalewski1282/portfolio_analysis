```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(42)
n_samples = 10000
mean = 0.1
std_dev = 0.2
a, b = -0.3, 0.3

samples = np.random.normal(loc=mean, scale=std_dev, size=n_samples)


pdf = norm.pdf(samples, loc=mean, scale=std_dev)
cdf_a = norm.cdf(a, loc=mean, scale=std_dev)
cdf_b = norm.cdf(b, loc=mean, scale=std_dev)
cdf_samples = norm.cdf(samples, loc=mean, scale=std_dev)


narrowed_pdf = pdf / (cdf_b - cdf_a)
narrowed_pdf[(samples < a) | (samples > b)] = 0


plt.figure(figsize=(8, 6))
bins = np.linspace(-1, 1, 100)


plt.hist(samples, bins=bins, alpha=0.6, label='Rozkład X', color='red', edgecolor='black', density=True)


plt.hist(samples, bins=bins, weights=narrowed_pdf, alpha=0.5, label='Zawężony rozkład X', color='blue', edgecolor='black', density=True)


plt.xlabel('X', fontsize=12)
plt.ylabel('Częstość', fontsize=12)
# plt.title('Porównanie rozkładu oryginalnego i zawężonego', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
