```python
import pandas as pd

df = pd.read_csv(r'./data/quarterly_return_rates.csv', index_col=0, parse_dates=True)

df
```

```python
rolling_mean_df = df.rolling(window=32).mean()
rolling_mean_df.dropna(inplace=True) # dropping first 7 rows as only next ones are calculated as we want full 8-quarter windows
# rolling_mean_df

positive_mean_stocks_df = rolling_mean_df[rolling_mean_df > 0].dropna(axis=1)

positive_mean_stocks_df
```

```python
positive_stocks_list = positive_mean_stocks_df.columns.to_list()
```

```python
filtered_df = df[positive_mean_stocks_df.columns]

filtered_df.to_csv(r'./data/positive_quarterly_return_rates.csv')

filtered_df.head()
```
